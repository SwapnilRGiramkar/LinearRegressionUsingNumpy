```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Size of House 
x=np.array([600,800,1000,1200,1800])
print(x)
```

```python
#Price of House
y=np.array([150,200,250,300,350])
print(y)
```

```python
#Normalize data
x=x/1000
print(x)
```

```python
#initializing paerameters
m=0
b=0
learning_rate=0.01
epochs=1000
n=len(x)
print(n)
```

```python
def compute_cost(x,y,m,b):
    predictions= m*x+b
    cost= (1/n)*np.sum((predictions-y)**2)
    return cost
```

```python
#Gradient descent training 
for i in range(epochs):
    predictions=m*x+b
    #gradient
    dm=(2/n)*np.sum((predictions-y)*x)
    db=(2/n)*np.sum(predictions-y)
    #Update parameters
    m=m-learning_rate*dm
    b=b-learning_rate*db
```

```python
if i%100==0 :
    cost=compute_cost(x,y,m,b)
    print(f"Epoch {i}, Cost: {cost}")
```

```python
#predictions
def predict(size):
    size=size/1000
    return m*size+b
    print("\n Model Trained !")
    print("Slope(m):",m)
    print("Intercept(b):",b)
```
